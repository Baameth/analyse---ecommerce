```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('ecom_marketing_real.db')
df = pd.read_sql_query("SELECT * FROM retail_brut", conn)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df['Mois'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df['CA'] =  df['UnitPrice']*df['Quantity']
CA_par_mois = df.groupby ('Mois')['CA'].sum()
```

```python
CA_par_mois
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(CA_par_mois.index.astype(str), CA_par_mois.values)
plt.xticks(rotation=45)
plt.title("Chiffre d'affaire par mois")
plt.xlabel('Mois')
plt.ylabel('CA')
```

```python
top_produits = df.groupby('Description')['Quantity'].sum()
top_10 = top_produits.sort_values(ascending=False).head(10)
top_10 = top_10.sort_values(ascending = True)
```

```python
plt.barh(top_10.index, top_10.values)
for i, (nom, valeur) in enumerate(zip(top_10.index, top_10.values)):
    plt.text(1000, i, nom, va='center', color='white', fontsize=8)
plt.xlabel('Quantité vendu')
plt.yticks([])
plt.title('Top 10 des produits')
plt.tight_layout()
plt.savefig('Top 10 des produits.png')
```

```python
import seaborn as sns

sns.barplot(CA_par_mois)
plt.xticks(rotation=45);
plt.title("Chiffre d'affaire par mois")
plt.xlabel('Mois');
plt.ylabel('CA');
```

```python
sns.lineplot(x=CA_par_mois.index.astype(str), y=CA_par_mois.values);
plt.xticks(rotation=45);
plt.title("Chiffre d'affaire par mois")
plt.xlabel('Mois');
plt.ylabel('CA');
```

```python
Pays_hors_UK = df[df['Country'] != 'United Kingdom'] 
CA_pays = Pays_hors_UK.groupby('Country')['CA'].sum()
CA_pays_top_10 = CA_pays.sort_values(ascending=False).head(10)
CA_pays_top_10
```

```python
sns.barplot(CA_pays_top_10);
plt.xticks(rotation=90);
plt.title('CA par pays')
```
